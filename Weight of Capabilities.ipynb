```python
# Import the standard packages of Python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Weight of Capabilities

```python
# Read in the csv file and create a dataframe
df = pd.read_csv('Weight of Capabilities.csv', sep=',', skiprows=6, error_bad_lines=False)

# Keep the desired columns
df = df[['[run number]','[step]', 'weighting_capabilities', 'covariance_cap_sec_aff','covariance_cap_sec_sus','covariance_cap_sec_aut',
         'covariance_cap_sec_com','covariance_cap_sec_inc','covariance_cap_aff_sus','covariance_cap_aff_aut',
         'covariance_cap_aff_com','covariance_cap_aff_inc', 'covariance_cap_sus_aut', 'covariance_cap_sus_com',
         'covariance_cap_sus_inc', 'covariance_cap_aut_com','covariance_cap_aut_inc','covariance_cap_com_inc']]

# Rename the columns
df = df.rename(columns={'[run number]':'Replication','[step]':'Step', 'weighting_capabilities':'Weighting-Option', 'covariance_cap_sec_aff':'Sec-Aff','covariance_cap_sec_sus':'Sec-Sus',
                        'covariance_cap_sec_aut':'Sec-Aut', 'covariance_cap_sec_com':'Sec-Com', 'covariance_cap_sec_inc':'Sec-Inc',
                        'covariance_cap_aff_sus':'Aff-Sus','covariance_cap_aff_aut':'Aff-Aut', 'covariance_cap_aff_com':'Aff-Com',
                        'covariance_cap_aff_inc':'Aff-Inc', 'covariance_cap_sus_aut':'Sus-Aut', 'covariance_cap_sus_com':'Sus-Com',
                        'covariance_cap_sus_inc':'Sus-Inc','covariance_cap_aut_com':'Aut-Com','covariance_cap_aut_inc':'Aut-Inc',
                        'covariance_cap_com_inc':'Com-Inc'})

# Keep only the rows of the last tick
df = df[df["Step"].apply(lambda x: (x > 4))]

# Show the first five rows of the dataframe
df.head()
```

```python
# Print the min, mean and max of the relation between capabilities security of supply and affordability 
min_cap = df['Sec-Aff'].min()
print(min_cap)

mean_cap = df['Sec-Aff'].mean()
print(mean_cap)

max_cap = df['Sec-Aff'].max()
print(max_cap)
```

```python
# Create a scatter plot of the relation between capabilities security of supply and affordability
# Grouped by the weighting option

fig, ax = plt.subplots()

colors = {1:'red', 2:'blue', 3:'green', 4:'black'}

grouped = df.groupby('Weighting-Option')
for key, group in grouped:
    group.plot(ax=ax, kind='scatter', x='Replication', y='Sec-Aff', label=key, color=colors[key], ylim=(-0.001,0.001))

plt.show()
```
